```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

This Dataset is Highly unbalanced as there is 99.7 % legit transactions and just 0.3% fraud transactions.

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal transactions and fraudaulent transactions

Number of Fraudaulent Transactions --> 492

```python
legit_sample = legit.sample(n = 492)
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training Data and Testing Data


```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

Training the model

```python
model.fit(X_train,Y_train)
```

Model Evaluation
Accuracy Score

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data : " ,training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on test data : ',test_data_accuracy)
```
